```python
class Vehicle:
    def __init__(self,mk,mdl,yr,mil):
        
        self.__make = mk
        self.__model = mdl
        self.__year = yr
        self.__mileage = mil
        
        
    def get_make(self):
        return self.__make
        
    def get_model(self):
        return self.__model
        
    def get_year(self):
        return self.__year
        
    def get_mileage(self):
        return self.__mileage
        
    def set_make(self,mk):
        self.__make = mk
        
    def set_model(self,mdl):
        self.__model = mdl
        
    def set_year(self,yr):
        self.__year = yr
        
    def set_mileage(self,mil):
        self.__mileage = mil
        
    def __str__(self):
        return "Make is :" + self.__make + "; Model is :" + self.__model + "; Year is :" + str(self.__year) + "; Mileage is :" + str(self.__mileage)
```

```python
class Employee:
    def __init__(self,epnm,epad,vd):
        self.__Emp_Name = epnm
        self.__Emp_Add = epad
        self.__Vehicle_Data = vd
    def get_Emp_Name(self):
        return self.__Emp_Name
    def get_Emp_Add(self):
        return self.__Emp_Add
    def get_Vehicle_Data(self):
        return self.__Vehicle_Data
    def set_Emp_Name(self,epnm):
        self.__Emp_Name = epnm
    def set_Emp_Add(self,epad):
        self.__Emp_Add = epad
    def set_Vehicle_Data(self,vd):
        self.__Vehicle_Data = vd
    def get_data_op4(self):
        if self.__Vehicle_Data.get_mileage()>78000:
            print('Employee name is: {0} {1};'.format(self.get_Emp_Name(),self.get_Vehicle_Data()))
            
    def compensation(self):
        pass
    def __str__(self):
        allData = "\nEmployee Name: " + self.__Emp_Name + "; Employee Address: " + self.__Emp_Add 
        allData += "\n" + self.__Vehicle_Data.__str__()
        return allData
```

```python
class Full_Time_Employee(Employee):
    def __init__(self,epnm,epad,vd,sal):
        super().__init__(epnm,epad,vd)
        self.__Salary = sal
        #self.__Taxes = tax
    def get_Salary(self):
        return self.__Salary
    def set_Salary(self,sal):
        self.__Salary = sal
    def compensation(self):
        if self.__Salary <= 40000:
            comp = self.__Salary - (0.16*self.__Salary)
        elif self.__Salary > 40000 and self.__Salary <= 75000:
            comp = self.__Salary - (0.16*40000) - ((self.__Salary-40000)*0.26)
        elif self.__Salary > 75000 :
            comp = self.__Salary - (0.16*40000) - (0.26*35000) - ((self.__Salary - 75000)*0.33)
        return "{0}'s compensation is {1:0.2f}".format(self.get_Emp_Name(),comp)
        
    def __str__(self):
        childData = super().__str__()
        childData = "\nDetails of the Employee are: " +childData
        childData += "\nSalary is " + str(format(self.__Salary,'0.2f')) 
        return childData
```

```python
class Hourly_Employee(Employee):
    def __init__(self,epnm,epad,vd,hw,hr):
        super().__init__(epnm,epad,vd)
        self.__Hours_Worked = hw
        self.__Hourly_Rate = hr
        #self.__Total = total
    def get_Hours_Worked(self):
        return self.__Hours_Worked
    def get_Hourly_Rate(self):
        return self.__Hourly_Rate
    def set_Hours_Worked(self,hw):
        self.__Hours_Worked = hw
    def set_Hourly_Rate(self,hr):
        self.__Hourly_Rate = hr
    def compensation(self):
        if self.__Hours_Worked <=40:
            comp = self.__Hours_Worked * self.__Hourly_Rate
        else:
            comp = (40*self.__Hourly_Rate) + ((self.__Hours_Worked - 40)*self.__Hourly_Rate*1.7)
        return '{0} is {1:0.2f}'.format(self.get_Emp_Name(),comp)
    def __str__(self):
        childData = super().__str__()
        childData = "\nDetails of the Employee are: " +childData
        childData += "\nHourly rate is " + str(self.__Hourly_Rate) + "\n Hours Worked is " + str(self.__Hours_Worked)
        return childData
```

```python
class Consultant(Employee):
    def __init__(self,epnm,epad,vd,ptype,hw):
        super().__init__(epnm,epad,vd)
        self.__Project_Type = ptype
        self.__Hours_Worked = hw
    def get_Project_Type(self):
        return self.__Project_Type
    def get_Hours_Worked(self):
        return self.__Hours_Worked
    def set_Hours_Worked(self,hw):
        self.__Hours_Worked = hw
    def set_Project_Type(self,ptype):
        self.__Project_Type = ptype
    def compensation(self):
        if self.__Project_Type == 1:
            comp = 55 * self.__Hours_Worked
        elif self.__Project_Type == 2:
            comp = 70 * self.__Hours_Worked
        elif self.__Project_Type == 3:
            comp = 85 * self.__Hours_Worked
        return '{0} is {1:0.2f}'.format(self.get_Emp_Name(),comp)

    def __str__(self):
        childData = super().__str__()
        childData += "\nDetails of the Employee are: " +childData
        childData += "\nProject Type: " + str(self.__Project_Type) + "\nHours Worked: " + str(self.__Hours_Worked)
        return childData
```

```python
import pickle
import sys

def main():
    run_menu_options()
def run_menu_options():
    employee_list = read_file_data()
    sel = 1
    while sel >=1 and sel <=5:
        print('\n===== Menu =====')
        print('1. To add Employee')
        print('2. To print Name and Address of all the Employees')
        print('3. To print Name and  Compensation of all Employees')
        print('4. To print Name with Vehicle Data for Vehicles with High Mileage')
        print('5. To exit')
        try:
            sel = int(input('Your selection: '))
            if sel < 1 and sel > 5:
                raise ValueError
        except ValueError:
            print('Enter valid interger')
            print('Try again')
            sel = 1
            continue
        if sel == 1:
            employee_list.append(run_option1())
        elif sel == 2:
            if len(employee_list) == 0:
                print('No employees in Database')
            else:
                run_option2(employee_list)
        elif sel == 3:
            if len(employee_list) == 0:
                print('No employee in Database')
            else:
                run_option3(employee_list)
        elif sel == 4:    
            if len(employee_list) == 0:
                print('No employee in Database')
            else:
                run_option4(employee_list)
        elif sel == 5:
            run_option5(employee_list)
def read_file_data():
    my_list = []
    try:
        fb = open('empdata1.dat','rb')
    except FileNotFoundError:
        return my_list
    else:
        try:
            while(True):
                my_list.append(pickle.load(fb))
        except EOFError:
            fb.close()
            return my_list
def run_option1():
    sel = 0
    while sel < 1 or sel > 4:
        try:
            sel = int(input('Type of Employee? (1.Full time 2. Hourly 3. Consultant)'))
        except ValueError:
            print('Incorrect Input')
            continue
    epnm,epad = get_employee_input()
    mk,mdl,yr,mil = get_vehicle_input()
    ob_vh = Vehicle(mk,mdl,yr,mil)
    if sel == 1:
        sal = get_fulltime_input()
        ob_emp = Full_Time_Employee(epnm,epad,ob_vh,sal)
    elif sel == 2:
        hw,hr = get_hourly_input()
        ob_emp = Hourly_Employee(epnm,epad,ob_vh,hw,hr)
    elif sel == 3:
        ptype,hw = get_consultant_input()
        ob_emp = Consultant(epnm,epad,ob_vh,ptype,hw)
    return ob_emp
    
def run_option2(my_list):
    for obj in my_list:
        print(obj)

def run_option3(my_list):
    print("\nEmployee Name and Compensation of all Employees")
    print("===================================================")
    for obj in my_list:
        print(obj.compensation())

def run_option4(my_list):
    print('Employee Name and Vehicle Data for High Mileage vehicles')
    print("==========================================================")
    for obj in my_list:
        obj.get_data_op4()

def run_option5(my_list):
    print('You choose to exit the program')
    yes_no = input('Do you want to exit?(Y/N)')
    if yes_no.lower() == 'y':
        fb = open('empdata1.dat','wb')
        for obj in my_list:
            pickle.dump(obj,fb)
        fb.close()
        sys.exit("Exiting the Program!")
        
def get_employee_input():
    e_name = input("Enter Employee's name ")
    e_addr = input("Enter Employee's address ")
    return e_name,e_addr
    
def get_vehicle_input():
    v_make = input("Enter vehicle's make ")
    v_model = input("Enter vehicle's model ")
    err_flag = True
    while err_flag:
        try:
            v_year = int(input("Enter vehicle's year "))
        except ValueError:
            print('\nEnter a numerical value for year')
        else:
            err_flag = False
    err_flag = True
    while err_flag :
        try:
            v_mileage = int(input("Enter vehicle's mileage "))
        except ValueError:
            print('\nEnter a numerical value for mileage')
        else:
            err_flag = False
    return (v_make,v_model,v_year,v_mileage)
    
def get_fulltime_input():
    err_flag = True
    while err_flag:
        try:
            f_sal = int(input("Enter salary for Full-time employee "))
        except ValueError: 
            print('\nEnter a numerical value for salary')
        else:
            err_flag = False
    return f_sal

def get_hourly_input():
    err_flag = True
    while err_flag:
        try:
            h_hw = int(input("Enter hours worked for Hourly employee "))
        except ValueError: 
            print('\nEnter a numerical value for hourly rate')
        else:
            err_flag = False
    
    err_flag = True
    while err_flag:
        try:
            h_hr = int(input("Enter rate for Hourly employee "))
        except ValueError: 
            print('\nEnter a numerical value for hours worked')
        else:
            err_flag = False
    return (h_hw,h_hr)

def get_consultant_input():
    err_flag = True
    while err_flag:
        try:
             c_ptype = int(input("Project type? (Enter a number between 1 and 3): "))
        except ValueError:
            print('\nEnter a numerical value for project type')
        else:
            err_flag = False
    
    err_flag = True
    while err_flag:
        try:
            c_hw = int(input("Enter hours worked for consultant "))
        except ValueError: 
            print('\nEnter a numerical value for hours worked')
        else:
            err_flag = False
    return (c_ptype, c_hw)

main()
```
